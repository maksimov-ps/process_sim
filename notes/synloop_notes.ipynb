## Steady state: conservation of mass

This chapter is focused on deriving a mass balance equation for $\frac{dy}{dz}$ for a non-isothermal and non-isobaric gas phase system.
$$ \ $$

A general form of a plug flow reactor mass balance conservation equation for a constant density system is taken from Perry's Chemical Engineers' Handbook (9th Edition) (eq. 19-17):

[amount introduced per unit of time] - [amount leaving per unit of time] - [reactant converted per unit of time] = [amount accumulated per unit of time]

$$ \frac{\partial (V' C)}{\partial V_r} - r(C, T) + \frac{\partial C}{\partial t} \tag{1} $$

where: 
- $V'$ - volumetric flowrate, [m^3/s]
- $C$ - molar concentration, [mol/m^3]
- $r(C, T)$ - reaction rate, [mol/s/m^3]
- $T$ - temperature, [K]
- $t$ - time, [s]


therefore: 
- $\frac{\partial (V' C)}{\partial V_r}$ - denotes [amount introduced per unit of time] - [amount leaving per unit of time]
- $r(C, T)$ - denotes [reactant converted per unit of time]
- $\frac{\partial C}{\partial t}$ - denotes [amount accumulated per unit of time]


allowing the following simplifications: 
- no transient phenomena and, therefore, no accumulation term
- no radial or axial dispersion, leading to a single spatial coordiante along the length of the plug flow reactor - $z$
- constant inlet flowrate of the reactants so that $V' = u_0$, where $u_0$ - superficial velocity [m/s]

and rewriting the rate equation considering the mass of catalyst, equation (1) can be rewritten as follows: 

$$ \frac{u_0}{\epsilon_{cat}} \frac{dC_i}{dz} - \rho_{cat} \sum_{i=1}^I\sum_{j=1}^J \nu_{i,j} r_j = 0 \tag{2} $$

where: 
- $u_0$ - superficial velocity, [m/s]
- $\epsilon_{cat}$ - porosity of the catalyst bed, [-]
- $C$ - molar concentration, [mol/m^3]
- $z$ - spatial coordinate along the length of the reactor, [m]
- $\rho_{cat}$ - bulk density of the catalyst bed, [kg_cat/m^3]
- $\nu_{i,j}$ - stoichiometric number of component $i$ for reaction $j$, 
- $r_j$ - reaction rate per unit of catalyst mass, [mol/s/kg_cat]
